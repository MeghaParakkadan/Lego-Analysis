## Legos Dataset Analysis
This project will allow us to do analysis on Legos dataset and let us anwer the below questions. 
1. What percentage of all licensed sets ever released were star wars themed.
2. In which year was Star warsnot not the most popular licensed theme( in terms of number of sets released that year)

There are 4 datasets given for analysis. They are
1. lego_sets.csv
2. parent_themes
3. sets
4. themes

First, we will look into the data set and have an idea about columns ,rows and what data it contain.

```python
import pandas as pd
df = pd.read_csv('Dataset/lego_sets.csv')
df.head(4)
```

```python
theme = pd.read_csv('Dataset/parent_themes.csv')
theme.head(10)
```

Lets combine both the table and get the data together. 

```python
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged=merged.drop(columns='name_y')
merged.head(10)
```

Now we got the entire data. To answer the first question, we have to filter out rows having is_licensed column value=False. 

```python
licensed = merged[merged['is_licensed']]
licensed.head(10)
```

set_num column is critical. So missing values are not acceptable.

```python
licensed = licensed.dropna(subset=['set_num'])
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
the_force = int((star_wars.shape[0]/licensed.shape[0])*100)
print(the_force)
```

51% of all licensed sets ever released were star wars themed. Now lets move on to the second question. 

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted ['count']= 1
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
max_df = summed_df.sort_values('count',ascending= False).drop_duplicates('year')
max_df.sort_values('year', inplace=True)

```

This will give us the 2017 as the answer for second question
